```python
#Min Max Normalization
x1= (x1 - x1.min()) / (x1.max() - x1.min())
x2= (x2 - x2.min()) / (x2.max() - x2.min())
y= (y - y.min()) / (y.max() - y.min())
```

```python
def h(a,x,b):
    z = (a*x) + b
    return 1 / (1+np.exp(-z))
```

```python
def error(a,x,b,y):
    
    error=0
    m=len(x)
    
    for i in range(m):
        error += np.power( h(a,x[i],b) - y[i], 2)
        
    return (1/2) * (error/m)
```

```python
def step_gradient(a,x,b,y,learning_rate):
    grad_a = 0
    grad_b = 0
    m = len(x)
    
    for i in range(m):
        grad_a += (1/m) * (h(a,x[i],b) - y[i]) * x[i]
        grad_b += (1/m) * (h(a,x[i],b) - y[i])
    
    a = a - (grad_a * learning_rate)
    b = b - (grad_b * learning_rate)
    
    return a,b
```

```python
def learn(init_a, x, init_b, y, learning_rate, iters):
    
    a,b = init_a, init_b
    
    for i in range(iters):
        a,b = step_gradient(a, x, b, y, learning_rate)
        e = error(a,x,b,y)
        all_errors.append(e)
        
        if i%1000 == 0:
            print("Errors:", e, "a:", a, "b:", b)
        
    return a,b
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain, xtest, ytrain, ytest = train_test_split( x2, y, test_size=0.2)
```

```python
final_a, final_b = learn(init_a, xtrain, init_b, ytrain, learning_rate, iters)
```

```python
def accuracy(a,xt,b,yt):
    t = h(a,xt,b)
    pred = []
    
    for i in t:
        if i > 0.5:
            pred.append(1.)
        else:
            pred.append(0.)
    
    print(pred)
    correct = 0
    
    m = len(yt)
    for i in range(m):
        if t[i] == yt[i]:
            correct += 1
            
    return correct / m
```
